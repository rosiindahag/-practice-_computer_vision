# **Histogram for a Tiny Randomly Generated Image**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generate 2-bit image
rand_img = np.random.uniform(low=0, high=3, size=(5,5))
rand_img
```

```python
rand_img = np.uint8(rand_img) #unsigned integer
rand_img
```

```python
hist = np.histogram(rand_img,bins=np.max(rand_img)+1)
plt.bar(x=np.unique(rand_img), height=hist[0], align="center", width=0.2)
plt.xticks([i for i in range(np.max(rand_img)+1)],fontsize=10)
plt.yticks(np.arange(0,14,2), fontsize=10)
plt.title("Image Histogram")
plt.xlabel("image value")
plt.ylabel("pixel count")
```

```python
plt.imshow(rand_img)
plt.colorbar()
```

# Extracting GLCM Features

```python
img_2=skimage.io.imread('23.JPG', as_gray=True)
img_2=np.uint8(img_2*255)
plt.imshow(img_2, cmap='gray')
```

```python
 glcm = skimage.feature.greycomatrix(img_2, distances=[6], angles=[0], levels=256, normed=True)
```

```python
dissimilarity = skimage.feature.greycoprops(P=glcm, prop='dissimilarity')
correlation = skimage.feature.greycoprops(P=glcm, prop='correlation')
homogeneity = skimage.feature.greycoprops(P=glcm, prop='homogeneity')
energy = skimage.feature.greycoprops(P=glcm, prop='energy')
contrast = skimage.feature.greycoprops(P=glcm, prop='contrast')
ASM = skimage.feature.greycoprops(P=glcm, prop='ASM')
```

```python
glcm_props = [dissimilarity, correlation, homogeneity, energy, contrast, ASM]
glcm_props_name = ['Dissimilarity', 'Correlation', 'Homogeneity', 'Energy', 'Contrast', 'ASM']
for i in range(len(glcm_props)):
  print(glcm_props_name[i],glcm_props[i])
```
